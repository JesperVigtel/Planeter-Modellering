### 2. Teori

#### 2.1 Newtons Gravitasjonslov
To legemer med masser m₁ og m₂ og avstand r, tiltrekker hverandre med en kraft F som er rettet langs den rette linjen mellom legemene, og som er proporsjonal med produktet av massene og omvendt proporsjonal med kvadratet av avstanden. Her er γ Newtons gravitasjonskonstant. Den har verdien 6,67 · 10⁻¹¹ m³kg⁻¹s⁻². 

$$ \vec{F} = \gamma \frac{m_1 \cdot m_2}{r^2} $$

##### 2.1.2 Lengde
Vi tar utgangspunkt i likningen for lengde under, som er $ r =\sqrt{(x-x_0)^2+(y-y_0)^2} $


Dette kan skrives ut i x og y retninger. (og z)


$$ F_x = −\gamma\frac{m_1·m_2·x(t)}{r^3} => −\gamma\frac{m_1·m_2·x(t)}{((x(t)−x_2(t))^2+(y(t)−y_2(t))^2)^\frac{3}{2}} $$

$$ F_y = −\gamma\frac{m_1·m_2·y(t)}{r^3} => −\gamma\frac{m_1·m_2·y(t)}{((x(t)−x_2(t))^2+(y(t)−y_2(t))^2)^\frac{3}{2}} $$


#### 2.2.1 Akselerasjon
Newtons andre lov sier at når et legeme blir påvirket av én eller flere krefter, vil det få en akselerasjon i den retningen kreftene virker. Summen av kreftene på legemet er lik legemets masse ganger dets akselerasjon.

Denne loven kan beskrives med formelen

 $$ \Sigma F = m_1 \cdot a  $$

der $\Sigma F$ er summen av kreftene som virker på legemet, $m_1$ er massen til legemet og a er akselerasjonen.

Deler vi på $m_1$ på begge sider, får vi at 

$$ \vec{a} = \gamma \frac{m_2}{r^2} $$

Dette kan skrives ut på x- og y-aksen som: 
$$ a_x(t) = \gamma \frac{m_2*x(t)}{r^3} $$
$$ a_y(t) = \gamma \frac{m_2*y(t)}{r^3} $$

hvor $a_x(t) \space og\space a_y(t)$ er akselerasjonen til planeten med masse $m_1$. 

#### 2.2.2 Fart og Posisjon 
For å finne posisjonene x(t) og y(t) må vi finne den deriverte av posisjonen (fart). Vi tar utgangspunkt i at den deriverte av posisjone er farten, og den deriverte av farten er akselerasjonen. Dette bruker vi i Eulers metode for å finne de neste verdiene. Vi tar utgangspunkt i at  $ a(t) = v'(t)$ og at $v(t) = s'(t) $. Da får vi at:
$$ v'_x(t) = \gamma \frac{m_2*x(t)}{r^3} \space og \space x'(t) = v_x(t) $$ og $$ v'_y(t) = \gamma \frac{m_2*y(t)}{r^3} \space og \space y'(t) = v_y(t)$$.


## 3 Modeller
____
### Modell 1 : Sola og Jorda

Nå skal vi se på et solsystem som består av kun jorda og sola. Modellen antar at sola står stille. Du kan derfor fokusere på hvordan jordas bane blir påvirket av sola.I denne modellen lar vi jordas masse $m_1=3·10^{−6}$ solmasser og solas masse $m_2 = 1 \space solmasse$. Ved å bruke AU (enhet brukt som den gjennomsnittlige avstanden mellom sol og jord) og år som enheter i vår modell, kan en finne at $\gamma =4π^2$.

La startbetingelsene til jorda være x(0) =1 ,y(0) =0, $v_x(0) =0$ og $v_y(0) = 2π$ .Sola kan stå i origo. Det betyr at i dette tilfellet vil $x_2(t) = 0 $ og $y_2(t) =0$for alle tider t.


#### Eulers metode
Vi bruker eulers metode for å finne farten og posisjonen. 
$ f(x+1) = f(x) + f'(x)*dx $. 
#### Euler-Cromers metode
I Euler-Cromers metode finner du det neste steget, ved å ta det forrige steget, + den deriverte av det neste steget.
$ f(x+1) = f(x) + f'(x+dx)*dx $.



### Modell 2: Med planeter
Vi utvider modellen ved å innføre flere planeter. Vi modellerer derfor for de fire innerste planetene til sola. 
Vi fortsetter å bruke Euler-Cromers metode for posisjonene, og bruker teori fra 2.2.1 til å finne akselerasjonspåvirkning mellom planeter.  

$$ a = \gamma \frac{m_2*(posisjon_{planet1}-posisjon_{planet2})}{r^3} $$


### 3. Resultater


##### Modell 1

```python
from pylab import *

# Variabler og konstanter
G = 4 * pi**2    #Gravitasjonskonstanten 𝛾 
m_1 = 3*1e-6     #Jordmassen i solmasser
solmasse = 1     #Solmassen i solmasser
x_0 = 1          #Startposisjon x-aksen
y_0 = 0          #Startposisjon y-aksen
vx_0 = 0         #Startfart x-aksen
vy_0 = 2*pi      #Startfart y-aksen

x2 = 0           #Posisjonen til sola
y2 = 0

#Parametre
N = 1000
a = 0
b = 1
t = linspace(a,b,N)
dt = t[1]-t[0]

#Arrayer
x = zeros(N+1)
y = zeros(N+1)
vx = zeros(N+1)
vy = zeros(N+1)


#Startverdier
x[0] = x_0
y[0] = y_0
vx[0] = vx_0
vy[0] = vy_0


for i in range(N):
    a_x = -G*solmasse*x[i]/((x[i]-x2)**2 + (y[i]-y2)**2)**(3/2)    #Utrykk for akselerasjon fra teori 2.2.1
    a_y = -G*solmasse*y[i]/((x[i]-x2)**2 + (y[i]-y2)**2)**(3/2)
    
    #Eulers metode:
    vx[i+1] = vx[i] + a_x*dt                                  #For farten
    vy[i+1] = vy[i] + a_y*dt
    
    x[i+1] = x[i] + vx[i]*dt                                  #For posisjonen
    y[i+1] = y[i] + vy[i]*dt


figure(figsize=(10,8))             #Endrer størrelsen til figuren
    
    
#Plot jordas bane rundt sola
plot(0,0,"y*", markersize =30 )    #Sola
plot(x,y)                          #Jordas bane
title("Jordas bane rundt sola (Eulers)")
xlabel("Avstand (AU)")
ylabel("Avstand (AU)")
grid()
show()
```

##### Modell 1 , Euler-Cromers metode

```python
for i in range(N): #Eulers metode:
    a_x = -G*solmasse*x[i]/((x[i]-x2)**2 + (y[i]-y2)**2)**(3/2)    #Utrykk fra teori 2.2.1
    a_y = -G*solmasse*y[i]/((x[i]-x2)**2 + (y[i]-y2)**2)**(3/2)
    
    vx[i+1] = vx[i] + a_x*dt              #For farten med Eulers metode
    vy[i+1] = vy[i] + a_y*dt
    
    x[i+1] = x[i] + vx[i+1]*dt            #For posisjonen med Euler-Cromers metode
    y[i+1] = y[i] + vy[i+1]*dt


figure(figsize=(10,8))   #Endrer størrelsen til figuren
    
    
#Plot jordas bane rundt sola
plot(0,0,"y*", markersize =30 )    #Sola
plot(x,y)    #Jordas bane
title("Jordas bane rundt sola (Euler-Cromer)")
xlabel("Avstand (AU)")
ylabel("Avstand (AU)")
grid()
show()


#Forskjellige akselerassjonsverdier
aksel = []
for i in range(0,101,20):
    a_x = -G*solmasse*x[i]/((x[i]-x2)**2 + (y[i]-y2)**2)**(3/2)    #Utrykk fra teori 2.2.1
    a_y = -G*solmasse*y[i]/((x[i]-x2)**2 + (y[i]-y2)**2)**(3/2)
    a = round(norm([a_x,a_y]),2)
    aksel.append(a)

print("Liste av akselerasjoner:")                 
print(aksel)
```

##### Modell 2:   Flere planeter   

```python
from pylab import *
#Parametre
b = 10
N = 10000
m_sol = 1
planeter = 0   #Antall planeter
pos_sol = array([0, 0])
G = 4*pi**2


def aks(pos_p1, pos_p2, m_p2):                   #Formel for akselerasjon mellom to planeter 
    r = norm(pos_p1-pos_p2)                      #Regner ut avstand mellom arrayer som skalarprodukt av planetvektorer x og y. 
    a = -G*m_p2*(pos_p1-pos_p2)/(r**3)           #Utrykk fra teori 2.2.1
    return a


#Leser av filen "planeter_data.dat"
with open("planeter_data.dat" , "r") as infile:  #Åpner filen slik at vi kan lese innholdet
    planeter = int(infile.readline())            #Finner antall planeter i filen
    pos = zeros([N,2,planeter])                  #Finner posisjonen til planetene i filen
    fart = zeros([N,2,planeter])                 #Finner farten ...
    masse = zeros(N)
    
    #Leser av parameterne til alle planeten i filen, og legger inn i matriser.
    for i,line in enumerate(infile):              #Leser av linjer og "parametere" langs linjen
        data = line.split()  
        x = float(data[1])                        #X-posisjonen til planet nummer "i"
        y = float(data[2])                        #y-posisjonen til planet nummer "i"
        vx = float(data[3])                       #Farten i x-retning til planet nummer "i"
        vy = float(data[4])                       #Farten i y-retning til planet nummer "i"
        m = float(data[5])                        #Massen til planet "i"
        
        pos[0, : ,i] = array([x,y])               #Lager posisjon til planeten
        fart[0, : ,i] = array([vx,vy])            #Farten til planeten
        masse[i] = m                              #Massen
        
        
for i in range(N-1):
    for j in range (planeter):
        pos_planet = pos[i,:,j]                   #Henter posisjonen til planeten
        fart_planet = fart[i,:,j]                 #Henter farten ...
        aks_planet = aks(pos_planet,pos_sol,m_sol)#Finner akselerasjonen fra sola
        
        for k in range (planeter):
            if k != j:                            #Finne påvirkninger fra de andre planetene
                pos_planet2 = pos[i,:,k]          #Finner posisjon til planet 2
                fart_planet2 = fart[i,:,k]        #Finner fart til planet 2
                aks_planet += aks(pos_planet,pos_planet2,masse[k])   #legger til akselerasjonspåvirkningen til akselerasjonen til planeten
                
                
        fart[i+1,:,j] = fart[i,:,j] + dt*aks_planet       #Euler-Cromers metode
        pos[i+1,:,j] = pos[i,:,j] + dt*fart[i+1,:,j]
    
navn = ['Jorda','Merkur','Venus','Mars']          # liste med navn for planeter

figure(figsize=(10,8))           #Endrer størrelsen til figuren
plot(0,0,"y*", markersize = 30)  #Sola i midten

for P in range(planeter):        #Plotter banene til plantene
    x = pos[:,0,P]
    y = pos[:,1,P]
    plot(x,y, label = navn[P]) 

    
title("Awesome epicsause stjernefigur + ellipser")
xlabel("Avstand (AU)")
ylabel("Avstand (AU)")

legend()
grid()
show()
```

#### Pygame-modell

```python
import pygame
#Parametre
b = 10
N = 10000

#Kjør forrige program først

pygame.init()

#Variabler for skjerm
høyde = 840
bredde = 840
Svart = (0,0,0)
ICE = (158,221,250)

vindu = pygame.display.set_mode((bredde,høyde))
vindu.fill(ICE)

pygame.display.set_caption("Awesome epicsause stjernefigur + ellipser (on ice)")

#Sola
sol_x = int(bredde/2)
sol_y = int(høyde/2)
sola = (sol_x,sol_y)    #Solas posisjon på skjermen
GUL = (255,255,0)
pygame.draw.circle(vindu, GUL , sola, 20)    #Tegner sola


#Planeter 
farger = [(95, 128, 174),(140, 140, 148),(237, 177, 81),(147, 72, 56)]  #Fine farger til hver planet
størrelser = [14,8,12,10]   #Passende størrelser

skrift = pygame.font.SysFont("monospace", 15) # For å vise tekst
tekst = "År: "


run = True
i = 0
yr = 0
år = int(round(N/b))
while run:
    # For å avslutte programmet
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            run = False
    #Skjermen
    vindu.fill(ICE)
    pygame.draw.circle(vindu, GUL , sola, 20)    #Tegner sola

    #Skriver tekst
    t = skrift.render(tekst+str(yr), 12, (255,255,255)) # Skriver tekst som forteller hvilker år simuleringen er på
    vindu.blit(t,(1,1)) # Plasser tekst

    #Tegner på planetene
    for p in range(planeter): # gå gjennom hver planet
        
        pos_x = pos[i+1,0,p]  #Henter x-kordinatet til planeten P ved tiden i+1
        pos_y = pos[i+1,1,p]  #Henter y-kordinatet til planeten P ved tiden i+1
        pos_p = (int(round(200*pos_x+bredde/2)),int(round(200*pos_y+høyde/2)))
        pygame.draw.circle(vindu,farger[p], pos_p, størrelser[p])    #Tegner på for planetene i posisjon i+1

    # År
    if i == år:
        år+=int(round(N/b))
        yr+=1
        
        
    if i+1 < N-1:    
        i += 1
    else:
        i = 0
        pos[0,:,:] = pos[N-1,:,:]
        fart[0,:,:] = fart[N-1,:,:]

    pygame.display.update()


pygame.quit()
```

### 4. Diskusjon



##### Modell 1

Modeller sier at posisjonen ved neste tidssteg er lik posisjonen ved forrige tidssteg + en viss endring som kommer fra farten og et lite tidssteg.

Det ser ut som at modellene går i en sirkelbanene, men dersom vi ser på listen med akselerasjoner i modell 1.2, ser vi at det er forskjellige akselerasjoner. Dersom det hadde vært en perfekt sirkelbane, skulle absoluttverdien av akselerasjonen vært konstant. Vi antar derfor at det er en ellipsebane, men at det er en tilnærmet sirkulær sirkelbane. Årsaken til dette kan være at det egentlig skal være to brennpunkter i ellipsebanene, hvor sola er den ene, men at de ligger så nærme i vårt tilefelle, at det nesten ikke utgjør en merkbar forskjell. 


I modellen brukte vi også både Euler og Euler-cromers for å finne posisjonen til planeten. Fra modellene viser det seg at Euler-Cromers er mer nøyaktig, og dermed bedre. I Eulers metode ser vi at start og sluttpunkt ikke møtes, og banen er ikke like rund. Vi testet for ulike verdier av N (altså antall steg og steglengde), og fant ut at Eulers metode ble riktig for en stor N. Euler-Cromer var mer robust, og fungerte med lavere N.

Vi forutsetter at det kun er én planet i solsystemet, og tar dermed ikke med påvirkningen fra de andre planetene på jorda. Vi antar også at sola står stile, og det modelleres kun for ett år og tar ikke med at vi kan bli truffet av en gigantisk meteoritt som kan endre jordas bane drastisk. 






##### Modell 2

I denne modellen legger vi til de fire innerste planetene, og at de virker på hverandre med krefter. Som i modell 1, ser det her også ut som sirkelbaner, men det er altså ellipser. Vi gjør samme antakelser og forutsetninger som i modell 1, bare for alle de fire planetene. Planetbanene gjelder kun i 2 dimensjoner, så de er ikke nøyaktige nok for virkeligheten. 



##### Modell 3
Modell 3 baserer seg på modell 2, og har derfor alle de samme forutsetningen og antakelsene. Igjen så er det kun en 2. dimensjonal figur, og dermed ikke så virkelighetsnær.
