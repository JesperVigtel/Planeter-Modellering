#### 2.1 Newtons Gravitasjonslov
To legemer med masser m₁ og m₂ og avstand r, tiltrekker hverandre med en kraft F som er rettet langs den rette linjen mellom legemene, og som er proporsjonal med produktet av massene og omvendt proporsjonal med kvadratet av avstanden. Her er γ Newtons gravitasjonskonstant. Den har verdien 6,67 · 10⁻¹¹ m³kg⁻¹s⁻². 

$$ \vec{F} = \gamma \frac{m_1 \cdot m_2}{r^2} $$

##### 2.1.2 Lengde
Vi tar utgangspunkt i likningen for lengde under, som er $ r =\sqrt{(x-x_0)^2+(y-y_0)^2} $


Dette kan skrives ut i x og y retninger. (og z)


$$ F_x = −\gamma\frac{m_1·m_2·x(t)}{r^3} => −\gamma\frac{m_1·m_2·x(t)}{((x(t)−x_2(t))^2+(y(t)−y_2(t))^2)^\frac{3}{2}} $$

$$ F_y = −\gamma\frac{m_1·m_2·y(t)}{r^3} => −\gamma\frac{m_1·m_2·y(t)}{((x(t)−x_2(t))^2+(y(t)−y_2(t))^2)^\frac{3}{2}} $$


#### 2.2.1 Akselerasjon
Newtons andre lov sier at når et legeme blir påvirket av én eller flere krefter, vil det få en akselerasjon i den retningen kreftene virker. Summen av kreftene på legemet er lik legemets masse ganger dets akselerasjon.

Denne loven kan beskrives med formelen

 $$ \Sigma F = m_1 \cdot a  $$

der $\Sigma F$ er summen av kreftene som virker på legemet, $m_1$ er massen til legemet og a er akselerasjonen.

Deler vi på $m_1$ på begge sider, får vi at 

$$ \vec{a} = \gamma \frac{m_2}{r^2} $$

Dette kan skrives ut på x- og y-aksen som: 
$$ a_x(t) = \gamma \frac{m_2*x(t)}{r^3} $$
$$ a_y(t) = \gamma \frac{m_2*y(t)}{r^3} $$

hvor $a_x(t) \space og\space a_y(t)$ er akselerasjonen til planeten med masse $m_1$. 

#### 2.2.2 Fart og Posisjon 
For å finne posisjonene x(t) og y(t) må vi ta utgangspunkt i at  $ a(t) = v'(t)$ og at $v(t) = s'(t) $. Når vi vet dette, kan vi skrive v'(t) som a(t) fra 2.2.1 (forrige oppgave). da får vi at:
$$ v'_x(t) = \gamma \frac{m_2*x(t)}{r^3} $$ og $$ v'_y(t) = \gamma \frac{m_2*y(t)}{r^3} $$.

### Eulers metode
Vi bruker eulers metode for å finne farten og posisjonen. 
$ f(x+1) = f(x) + f'(x)*dx $. 
#### Euler-Cromers metode
$ f(x+1) = f(x) + f'(x+dx)*dx $.

##### Modell 1

```python
from pylab import *

# Variabler og konstanter
G = 4 * pi**2    #Gravitasjonskonstanten 𝛾 
m_1 = 3*1e-6     #Jordmassen i solmasser
solmasse = 1     #Solmassen i solmasser
x_0 = 1          #Startposisjon x-aksen
y_0 = 0          #Startposisjon y-aksen
vx_0 = 0         #Startfart x-aksen
vy_0 = 2*pi      #Startfart y-aksen

x2 = 0          #Posisjonen til sola
y2 = 0

#Parametre
N = 1000
a = 0
b = 1
t = linspace(a,b,N)
dt = t[1]-t[0]

#Arrayer
x = zeros(N+1)
y = zeros(N+1)
vx = zeros(N+1)
vy = zeros(N+1)


#Startverdier
x[0] = x_0
y[0] = y_0
vx[0] = vx_0
vy[0] = vy_0


for i in range(N): #Eulers metode:
    a_x = -G*solmasse*x[i]/((x[i]-x2)**2 + (y[i]-y2)**2)**(3/2)    #Utrykk fra teori 2.2.1
    a_y = -G*solmasse*y[i]/((x[i]-x2)**2 + (y[i]-y2)**2)**(3/2)
    
    vx[i+1] = vx[i] + a_x*dt                                  #For farten
    vy[i+1] = vy[i] + a_y*dt
    
    x[i+1] = x[i] + vx[i]*dt                                  #For posisjonen
    y[i+1] = y[i] + vy[i]*dt


figure(figsize=(8,8))
    
    
#Plot jordas bane rundt sola
plot(0,0,"y*", markersize =30 )    #Sola
plot(x,y)    #Jordas bane
title("Jordas bane rundt sola (Eulers)")
xlabel("Avstand (AU)")
ylabel("Avstand (AU)")
show()
```

##### Modell 1 , Euler-Cromers metode

```python
for i in range(N): #Eulers metode:
    a_x = -G*solmasse*x[i]/((x[i]-x2)**2 + (y[i]-y2)**2)**(3/2)    #Utrykk fra teori 2.2.1
    a_y = -G*solmasse*y[i]/((x[i]-x2)**2 + (y[i]-y2)**2)**(3/2)
    
    vx[i+1] = vx[i] + a_x*dt                                  #For farten
    vy[i+1] = vy[i] + a_y*dt
    
    x[i+1] = x[i] + vx[i+1]*dt                                  #For posisjonen
    y[i+1] = y[i] + vy[i+1]*dt


figure(figsize=(8,8))
    
    
#Plot jordas bane rundt sola
plot(0,0,"y*", markersize =30 )    #Sola
plot(x,y)    #Jordas bane
title("Jordas bane rundt sola (Euler-Cromer)")
xlabel("Avstand (AU)")
ylabel("Avstand (AU)")
show()
```

##### Modell 1
Modellen sier at antall smittede individer ved neste tidssteg er lik antall smittede individer ved forrige tidssteg + en viss andel (a) av antall individer som sprer smitten videre.

Modellen forutsetter enten at det ikke er noen immunitet innenfor smittemengden, altså at ingen smittede møter på personer som allerede er smittet. Det kan være et realistisk bilde i en stor populasjon i begynnelsen av et smitteforløp. Det er derfor usannsynlig at modellen beskriver utvikling langt fram i tid. Modellen forutsetter også at ingen blir friske i løpet av den tiden vi ser på. Igjen peker dette på at modellen kun kan gjelde for et kort tidsrom.
